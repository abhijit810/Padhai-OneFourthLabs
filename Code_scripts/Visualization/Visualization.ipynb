#Tabulation

```
url = 'https://api.covid19india.org/states_daily.json'
import urllib.request as request
import json
request.urlretrieve(url, 'covid_data.json');
```

```
covid_data = pd.read_json('covid_data.json')
```

```
with open('covid_data.json') as f:
  data = json.load(f)
data = data['states_daily']
```

```
covid_data = pd.json_normalize(data)
```

```
cd = covid_data
```

```
cd.date = pd.to_datetime(cd.date)
```

```
cd = cd[cd['status']=='Confirmed']
```

```
cd
```

```
cd.drop('status', axis=1, inplace=True)
```

```
cd.set_index('date',inplace=True)
```

```
cd.info()
```

```
cd.drop('dateymd', axis=1, inplace=True)
```

```
cd.apply(pd.to_numeric)
```

```
df = cd.tail()
```

```
df = df.apply(pd.to_numeric)
```

```
df.info()
```

```
def colour_red_neg(x):
  color = 'red' if x < 100 else 'blue'
  return 'color : ' + color
```

```
df.style.applymap(colour_red_neg)
```
