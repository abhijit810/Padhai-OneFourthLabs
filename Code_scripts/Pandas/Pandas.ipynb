### Task 1 : collect diameters and mass for each planet in solar system and compute the density

```
mass = pd.Series (
                  data = (0.330, 4.87, 5.97, 0.073, 0.642, 1898, 568, 86.8, 102, 0.0146) , 
                  index = ('MERCURY','VENUS','EARTH','MOON','MARS','JUPITER','SATURN','URANUS','NEPTUNE','PLUTO' )
                 )
mass['PlanetX'] = 6
mass_in_kg = mass * (10**24)
```

```
diameter = pd.Series (
                  data = (4879, 12104,	12756,	3475,	6792,	142984,	120536,	51118,	49528,	2370) , 
                  index = ('MERCURY','VENUS','EARTH','MOON','MARS','JUPITER','SATURN','URANUS','NEPTUNE','PLUTO' )
                 )

radius = diameter/2
```

```
radius
```

```
volume = (4/3) * pi * (radius**3)
```

```
density = mass_in_kg/volume
```

```
density # kg/km3
```

### Task 3 : Compare Dictionaries with series :
- a. checking if a key is present
- b. summing values
- c. computing std deviation

```
# creating a dictionary
%%time
dict_data = {}
N = 100000
for i in range(N):
  dict_data[i] = i%10
```

```
# creating a Series
%%time
series_data = pd.Series(dict_data)
```

a - checking if a key is present

```
#a -  checking if a key is present in dict
%%time
random_int = random.randint(0,N)
print('random_int',random_int)
if random_int in dict_data.keys() : print ('found')
else : print('not found')
```

```
#a -  checking if a key is present in series
%%time
random_int = random. randint(0,N)
print('random_int',random_int)
if random_int in series_data.index : print ('found')
else : print('not found')
```

b - summing values

```
#b - summing values in dict
%%time
sum = 0
for value in dict_data.values():
  sum = sum + value
print(sum)
```

```
#b - summing values in series
%%time
sum = 0
sum = np.sum(series_data)
```

c : computing standard deviation

```
#c - Computing standard_deviation in dictionary
%%time
sum = 0
for value in dict_data.values():
  sum = sum + value
print(sum)
mean = sum / len(dict_data.keys())
variance = 0
print('mean',mean)
deviation = {}
for key in dict_data.keys() :
  deviation[key] = dict_data[key] - mean
  variance = variance + (deviation[key]**2)
variance = variance /(len(dict_data.keys()) -1)
std_deviation_dict = (variance)**0.5
```

```
std_deviation_dict
```

```
#c - Computing standard_deviation in series
%%time
std_deviation_series = np.power(( np.sum( np.power((series_data - np.mean(series_data)),2) ) ) / (np.size(series_data) -1),0.5)
```

```
std_deviation_series
```

```
%%time
np.std(series_data)
```

```
def create_df(nrows,ncols):
  row_names = pd.Series(np.arange(1,nrows+1))
  row_names = list(map(lambda x : 'r' + str(x) , row_names ))

  col_names = pd.Series(np.arange(1,ncols+1))
  col_names = list(map(lambda x : 'c' + str(x), col_names ))
  print(row_names)
  print(col_names)
  df = pd.DataFrame(np.random.randint(1,nrows*ncols, (nrows,ncols)) )
  df.index = row_names
  df.columns = col_names
  return df
```
