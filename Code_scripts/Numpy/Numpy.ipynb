## Exercise 1 : Area of circle/sphere quarter in a square/cube

```
N = 10000000
```

### 1a : Area of the quarter circle 2D

```
array_one = np.random.rand(N,2)
print(array_one)
```

```
hypotenuse = lambda base,height : (base**2 + height**2)**0.5
```

```
#test if the function is working
print(hypotenuse(3,4))
```

```
%%time
dist_map = np.array(list(map(hypotenuse, array_one[:,0],array_one[:,1])))
print(dist_map)
```

```
ratio = (dist_map[dist_map <= 1].size)/N
print(ratio)
```

### 1b : Area of a quarter sphere 3D

```
array_two = np.random.rand(N,3)
print(array_two)
```

```
distance_3d = lambda x,y,z : ( x**2 + y**2 + z**2 )**0.5 
```

```
dist_map_3d = np.array(list(map(distance_3d, array_two[:,0],array_two[:,1],array_two[:,2])))
```

```
ratio_3d = (dist_map_3d[dist_map_3d <= 1].size)/N
print(ratio_3d)
```

###1c : achieve the above in one single function line

```
def area_outside_circle(npoints, ndim):
    points = np.random.rand(npoints, ndim)
    return np.sum(np.sum(points * points, axis = 1) > 1)/npoints
```

```
%%time
area_outside_circle(N,10)
```
